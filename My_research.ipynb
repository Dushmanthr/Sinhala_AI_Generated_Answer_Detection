```python
df = pd.read_excel("Processed.xlsx")
df.head(5)
```

```python
df['Answers'] = df['Answers'].str.replace('*', '', regex=False)
```

```python
df['Answers'][1190]
```

```python
# Create an instance of SinhalaTokenizer
tokenizer = SinhalaTokenizer()


# Apply the tokenize method to the entire column
df['Tokenized_Answers'] = df['Answers'].apply(lambda answer: tokenizer.tokenize(sentence=str(answer)))
```

```python
df['Tokenized_Answers']
```

```python
df.to_excel("tokenized_answers.xlsx", index=False)
```

```python
df1 = pd.read_excel("tokenized_answers.xlsx")
df1.head(5)
```

```python
from sinling import SinhalaStemmer

stemmer = SinhalaStemmer()

df1['Stem_answers'] = df1['Tokenized_Answers'].apply(lambda tokenized_answer: stemmer.stem(word=str(tokenized_answer)))

```
